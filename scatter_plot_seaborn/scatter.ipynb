Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Automobile cars dataset from Seaborn

```python
df = sns.load_dataset("mpg")
df
```

Univariate Scatter Plot

```python
# create constant y value for all the data points (len(df['mpg']))
constant_y = [0] * len(df['mpg'])
```

```python
constant_y
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(x = df['mpg'], y = constant_y, 
                label='mpg')
plt.xlabel('mpg')
plt.title('Distribution of mpg')
plt.grid()
plt.show()
```

```python
# By observing this chart, we can say that more data points are overlapping on each other
# between 14 to 38 value of MPG
```

Bivariate Scatter Plot

```python
plt.figure(figsize=(8,3))
sns.scatterplot(x = df['horsepower'], y = df['acceleration'], 
                label='horsepower vs. acceleration')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.title('horsepower vs. acceleration')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# By observing this chart, we can say that horsepower and acceleration having
# low negative corelation
```

Multivariate Scatter Plot

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data = df, x='horsepower', y='acceleration', 
                hue='origin')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.title('horsepower vs. acceleration Scatter Plot (Color-coded by origin)')
plt.grid(True)
plt.legend(title='origin')
plt.show()
```

```python
# By observing this chart, we can say that 
# 1. for USA origin cars, horsepower and acceleration are having low negative corelation
# 2. for Japan origin cars, horsepower and acceleration are also having low negative corelation
# 3. for Europe origin cars, horsepower and acceleration are also having low negative corelation
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data = df, x='horsepower', y='acceleration', 
                size='origin', sizes=(100, 10))
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.title('horsepower, acceleration, and origin Scatter Plot')
plt.grid()
plt.show()
```

```python
# By observing this chart, we can say that 
# 1. for USA origin cars, horsepower and acceleration are having low negative corelation
# 2. for Japan origin cars, horsepower and acceleration are also having low negative corelation
# 3. for Europe origin cars, horsepower and acceleration are also having low negative corelation
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data=df, x='horsepower', y='acceleration',  
                size='origin', sizes=(100, 10), hue='origin')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.title('horsepower, acceleration, and origin Scatter Plot')
plt.grid()
plt.show()
```

```python
# By observing this chart, we can say that 
# 1. for USA origin cars, horsepower and acceleration are having low negative corelation
# 2. for Japan origin cars, horsepower and acceleration are also having low negative corelation
# 3. for Europe origin cars, horsepower and acceleration are also having low negative corelation
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data = df, x='horsepower', y='acceleration', 
                style='origin', markers=['o', '*','s'])
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.title('horsepower, acceleration, and origin Scatter Plot')
plt.grid()
plt.show()
```

```python
# By observing this chart, we can say that 
# 1. for USA origin cars, horsepower and acceleration are having low negative corelation
# 2. for Japan origin cars, horsepower and acceleration are also having low negative corelation
# 3. for Europe origin cars, horsepower and acceleration are also having low negative corelation
```
