Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Load dataset

```python
df1 = sns.load_dataset("mpg")
df1
```

```python
df1.columns
```

```python
df1.info()
```

Distribution plot with Cylinders column (Univariate)

```python
# Distribution plot represents how data is distributed at certian intervals
# To create dist plot, use sns.displot() function
```

```python
sns.displot(data=df1, x='cylinders', bins = 5)
plt.xlabel("Number of Cylinders")
plt.show()
```

```python
# Highest numbers of cars are using 4 to 5 cylinders
```

```python
# To display Kernel Density Estimate line along with Histogram, we can use kde = True
```

```python
sns.displot(data=df1, x='cylinders', bins = 5, kde = "True")
plt.xlabel("Number of Cylinders")
plt.show()
```

```python
# kde is smooth line that shows overall distribution of data
```

```python
# If we want to display only Kernel Density Estimate line without
# Histogram, we can use kind = kde
```

```python
sns.displot(data=df1, x='cylinders', kind = "kde")
plt.xlabel("Number of Cylinders")
plt.grid()
plt.show()
```

```python
# If we want to proportional densities, then we can use kind = "ecdf"
```

```python
sns.displot(data=df1, x='cylinders', kind = "ecdf")
plt.xlabel("Number of Cylinders")
plt.grid()
plt.show()
```

Distribution Plot for Bivariate Analysis

```python
# We can show relationship between 2 variables using dist plot
# Lets plot distribution plot for no of cylinders on x axis and mileage on y axis
```

```python
sns.displot(data=df1, x='cylinders', y='mpg')
plt.xlabel("Number of Cylinders")
plt.ylabel("mpg")
plt.show()
```

```python
# From graph we can say that maximum mileage given by cars has 4 cylinders
```

```python
# To display Kernel Density Estimate line , we can use kind = 'kde'
```

```python
sns.displot(data=df1, x='cylinders', y='mpg', kind='kde')
plt.xlabel("Number of Cylinders")
plt.ylabel("mpg")
plt.show()
```

```python
# From graph we can say that maximum mileage given by cars has 4 cylinders
```
