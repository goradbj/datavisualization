```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("mpg")
df
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.lineplot(data = df, x = 'horsepower', y='acceleration')

plt.subplot(2,2,2)
sns.lineplot(data = df, x = 'weight', y='mpg')

plt.subplot(2,2,3)
sns.boxplot(data = df, x = 'origin', y='weight')

plt.subplot(2,2,4)
sns.scatterplot(data = df, x = 'horsepower', y='acceleration')
plt.savefig("output.png")
plt.show()
```
