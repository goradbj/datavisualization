```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("products.csv", index_col = 0)
df
```

```python
df.categories
```

Vertical Bar Plots

```python
plt.figure(figsize=(6,3))
plt.bar(df.categories, df.total_manufactured_product, label = 'manufactured_product')
plt.xlabel("categories")
plt.ylabel("total_manufactured_product")
plt.title("categories vs total_manufactured_product")
plt.grid()
plt.legend()
plt.show()
```

Horizontal Bar Plots

```python
plt.figure(figsize=(6,3))
plt.barh(df.categories, df.total_manufactured_product, label = 'manufactured_product')
plt.ylabel("categories")
plt.xlabel("total_manufactured_product")
plt.title("categories vs total_manufactured_product")
plt.grid()
plt.legend()
plt.show()
```

Multiple Bar Plots

```python
plt.figure(figsize=(6,3))
x = np.arange(len(df.categories))
width = 0.35

plt.bar(x - width/2, df.tested_products, width, label='tested_products')
plt.bar(x + width/2, df.non_tested_products, width, label='non_tested_products')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("tested_products vs non_tested_products")
plt.xticks(x, df.categories)
plt.grid()
plt.legend()
plt.show()
```

Stacked Plots

```python
plt.figure(figsize=(6,3))
plt.bar(df.categories, df.tested_products, label='tested_products')
plt.bar(df.categories, df.non_tested_products, bottom=df.tested_products, label='non_tested_products')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("tested_products vs non_tested_products")
plt.xticks(x, df.categories)
plt.grid()
plt.legend()
plt.show()
```
