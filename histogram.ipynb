```python
import matplotlib.pyplot as plt
```

Symmetric

```python
ages = [25, 45, 35, 40, 42, 45, 47, 50, 51, 53, 55, 57, 60, 62, 65, 67, 70, 75, 80]
num_bins = 5
plt.figure(figsize=(8,3))
plt.hist(ages, bins=num_bins, edgecolor='k', alpha=0.7, color='orange', label="ages")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution Histogram')
plt.legend()
plt.show()
```

Bimodel

```python
ages = [5, 25, 15, 20, 22, 25, 27, 30, 31, 33, 35, 37, 40, 42, 45, 47, 50, 55, 60,
        85, 95, 85,90,92, 95, 97, 100, 101, 103, 105, 107, 110, 112, 115, 117, 120, 125, 130]
num_bins = 10
plt.figure(figsize=(8,3))
plt.hist(ages, bins=num_bins, edgecolor='k', alpha=0.3, color='blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution Histogram')
plt.show()
```
