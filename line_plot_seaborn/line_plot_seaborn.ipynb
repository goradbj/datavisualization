Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Dataset

```python
df = pd.read_csv("flights.csv")
df
```

Convert Month from Text to Numbers

```python
month_to_num = {
    "January": 1, 
    "February": 2,
    "March": 3,
    'April': 4, 
    'May': 5, 
    'June': 6, 
    'July': 7,
    'August': 8, 
    'September': 9, 
    'October': 10, 
    'November': 11, 
    'December':12
}
```

```python
# create new column as month_num by mapping above dict to month column
df['month_num'] = df['month'].map(month_to_num)
df
```

```python
# create new date column using year, month and day (01)
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month_num'].astype(str) + '-01')
df
```

1. What is overall passengers trend in last 12 years?

```python
plt.figure(figsize=(10, 3))
sns.lineplot(data=df, x='date', y='passengers')
plt.title('Passenger Counts Over the period')
plt.xlabel('date')
plt.ylabel('Passenger Count')
plt.grid()
plt.show()
```

```python
# By Analyzing this chart, we can say that number of passengers are increasing 
# over the period. Also there is seasonality
# trend is positive
```

2. tell me in which months more number of passengers are travelling?
3. tell me in which months demand is low, so we can give holidays to few staff members?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='month_num', y='passengers', hue='year')
plt.title('Passenger Counts Over Months')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.grid()
plt.show()
```

```python
# 2. May, June, July, Aug, Sept
# 3. Jan to Apr and Oct to Dec
```
