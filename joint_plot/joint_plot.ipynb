Import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Load dataset

```python
df = sns.load_dataset("penguins")
df
```

```python
df.island.unique()
```

Find out the relation between flipper length and body mass

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g')
plt.grid()
plt.show()
```

```python
# By Analyzing this graph, we can say that,
# 1. Scatter plot - as the flipper length increases, body mass increases
# 2. Flipper length hist- More penguins has flipper length between 180 to 205 & 210 to 225
# 3. Body mass hist- More penguins have body mass between 3000 to 5000
```

Can we show margins / axis to the histogram

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g',
              marginal_ticks=True)
plt.grid()
plt.show()
```

Can we change the Marker of data points

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g',
              marginal_ticks=True, marker="+")
```

Show the relation between flipper length and body mass with respect to specis

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g',
              marginal_ticks=True, hue="species")
plt.grid()
plt.show()
```

```python
# By Analyzing this graph, we can say that,
# Gentoo penguins have more flipper length & Body Mass than other two
```

Show the relation between flipper length and body mass with respect to specis with KDE

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g',
              marginal_ticks=True, hue="species", kind='kde')
plt.grid()
plt.show()
```

```python
# By Analyzing this graph, we can say that,
# Gentoo penguins have more flipper length & Body Mass than other two
```

Show the relation between flipper length and body mass with trend

```python
sns.jointplot(data=df, x='flipper_length_mm', y='body_mass_g',
              marginal_ticks=True, kind='reg')
plt.grid()
plt.show()
```

```python
# By Analyzing this graph, we can say that,
# Flipper length and Body Mass are positively coorelated
# if Flipper length increases, Body mass increases and vice versa.
```
