```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("products.csv", index_col = 0)
df
```

Vertical Bar Plot

```python
plt.figure(figsize=(6,3))
plt.bar(df.categories, df.total_manufactured_product, label = 'manufactured_products')
plt.xlabel('Categories')
plt.ylabel('total_manufactured_product')
plt.title('Categories vs total_manufactured_product')
plt.legend()
plt.show()
```

Horizontal Bar Plot

```python
plt.figure(figsize=(6,3))
plt.barh(df.categories, df.total_manufactured_product, label = 'manufactured_products')
plt.xlabel('Num_of_manufactured_products')
plt.ylabel('Categories')
plt.title('Categories vs manufactured_products')
plt.legend()
plt.show()
```

Multiple Bar Plots

```python
plt.figure(figsize=(6,3))
x = np.arange(len(df.categories))
width = 0.35
plt.bar(x - width/2, df.tested_products, width, label='tested_products')
plt.bar(x + width/2, df.non_tested_products, width, label='non_tested_products')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('tested_products vs non_tested_products')
plt.xticks(x, df.categories)
plt.legend()
plt.show()
```

Stacked Bar Plots

```python
plt.figure(figsize=(6, 3))
plt.bar(df.categories, df.tested_products, label='tested_products')
plt.bar(df.categories, df.non_tested_products, bottom=df.tested_products, label='non_tested_products')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('tested_products vs non_tested_products')
plt.legend()
plt.show()
```
